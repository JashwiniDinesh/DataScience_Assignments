```python
#load the libraries
import pandas as pd
import seaborn as sns
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
import numpy as np
from sklearn import linear_model
```

```python
data= pd.read_csv('D:/Data_ Science/Toyoto_Corrola.csv')
data.head()
```

```python
data.info()
```

```python
data1=data.drop('Model',axis = 1)
data1
```

```python
data2=data1.drop('Id',axis = 1)
data2
```

```python
df=data2.rename({'Age_08_04':'Age'},axis=1)
df
```

```python
#Build model
model = smf.ols('Price~Age+KM+HP+Doors+Cylinders+Gears+Weight',data=df).fit()
```

```python
#Coefficients
model.params
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)
```

```python
model.summary()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Age','KM','HP','Doors','Cylinders','Gears','Weight']],df.Price)
```

```python
reg.coef_
```

```python
reg.intercept_
```
